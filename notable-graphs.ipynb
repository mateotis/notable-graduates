```python
# linear regression with subset of Df_gradsavg
top200 = Df_gradsavg[Df_gradsavg['Rank'] <= 200]

# plotting
plot = sns.regplot(x='Rank', y='Average rank of graduates', data=top200)
plt.xlabel('THE World University Rankings 2020')

# regression stats
x = np.array(top200['Rank']).reshape((-1, 1))
y = np.array(top200['Average rank of graduates'])    # need to divide by age of university
x = sm.add_constant(x)
model = sm.OLS(y, x)
results = model.fit()
results.summary()
```

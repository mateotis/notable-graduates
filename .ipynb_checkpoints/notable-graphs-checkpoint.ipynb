```python
# notable-graphs: graphs, regressions, and the rest of the data analysis

import pandas as pd
pd.set_option('display.max_columns', 25000)
pd.set_option('display.max_rows', 25000)
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn')
import statsmodels.api as sm
```

```python
Df_THE = pd.read_csv("THE2020.csv",sep=',')
Df_merged = pd.read_csv("merged.csv",sep=',')
Df_grads = pd.read_csv("uniGrads.csv",sep=',')
Df_gradsstats = pd.read_csv("uniGradsWithStats.csv",sep=',')
Df_top200unis = pd.read_csv("top200unis.csv",sep=',')
```

```python
# checking whether rankings within a university are normally distributed
search = "University of Zagreb"
attendees = Df_merged[Df_merged['education'].str.contains(search) == True]
x = np.array(attendees['ranking_final_b_5'])

# import math
# x = [math.log(y,10) for y in x]
plt.figure(figsize=(10,7))
plt.xlabel('Ranking of Notables attending %s' % (search))
plt.ylabel('Proportion of Notables in Given Ranking')
plt.figtext(.3,.9,"Distribution of Notables from %s" % (search),fontsize=15)

sns.distplot(x)
```

```python
# lower mean or median of university notables is better, since lower notability value is better

# linear regression with subset of Df_top200unis
top200 = Df_top200unis[Df_top200unis['Rank'] <= 200]
# top200 = top200[top200['Log median rank / Age'] < 0.4]
column_for_x = 'Rank'
column_for_y = 'Log median rank / Age'

# plotting
plt.figure(figsize=(12,7))
plt.rc('xtick', labelsize=15)
plt.rc('ytick', labelsize=15)
plot = sns.regplot(x=column_for_x, y=column_for_y, data=top200, label='y = 0.0006x + 0.0620')
plt.xlabel('THE World University Rankings 2020')
plt.title("Ln(Median Rank of Graduates) divided by University Age vs. Rank of University (THE)",fontsize=22)
plt.xlabel(column_for_x, fontsize=19)
plt.ylabel("ln(median rank)/age", fontsize=19)
plt.legend(fontsize=13, loc='upper left')

# regression stats
x = top200[column_for_x]
y = top200[column_for_y]    # need to divide by age of university
x = sm.add_constant(x)
model = sm.OLS(y, x)
results = model.fit()
results_list = results.summary()
results_list

# print(top200[top200['Log median rank / Age'] >= 1.2])
```

```python
# residual of regression
Y_pred = results.predict(x)
residual = y.values-Y_pred
plt.scatter(x[column_for_x],residual)
plt.xlabel(column_for_x)
plt.ylabel("residual of %s" % (column_for_x))
plt.title("Residual Analysis of Rankings")
plt.show()
```

```python
# lower mean or median of university notables is better, since lower notability value is better

# linear regression with subset of Df_top200unis
top200_10 = Df_top200unis[(Df_top200unis['Rank'] <= 200) & (Df_top200unis['Number of graduates'] >= 10)]
# top200 = top200[top200['Log median rank / Age'] < 0.4]
column_for_x = 'Rank'
column_for_y = 'Log median rank / Age'

# plotting
plt.figure(figsize=(12,7))
plt.rc('xtick', labelsize=15)
plt.rc('ytick', labelsize=15)
plot = sns.regplot(x=column_for_x, y=column_for_y, data=top200_10, label='y = (7.691e-05)x + 0.0898')
plt.xlabel('THE World University Rankings 2020')
plt.title("Ln(Median Rank of Graduates) divided by University Age vs. Rank of University (THE), 10+ Graduates",fontsize=22)
plt.xlabel(column_for_x, fontsize=19)
plt.ylabel("ln(median rank)/age", fontsize=19)
plt.legend(fontsize=13, loc='upper right')

# regression stats
x_10 = top200_10[column_for_x]
y_10 = top200_10[column_for_y]    # need to divide by age of university
x_10 = sm.add_constant(x_10)
model = sm.OLS(y_10, x_10)
results = model.fit()
results_list = results.summary()
print(results_list)
```

```python
# residual of regression
Y_pred_10 = results.predict(x_10)
residual = y.values-Y_pred_10
plt.scatter(x_10[column_for_x],residual)
plt.xlabel(column_for_x)
plt.ylabel("residual of %s" % (column_for_x))
plt.title("Residual Analysis of Rankings")
plt.show()
```

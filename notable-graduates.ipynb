```python
uniList = open("unis.txt","r")
lines = uniList.read().splitlines()

unis = []
for uni in lines:
    print(uni)
    unis.append(uni)
```

```python
gradsDict = {}

for uni in unis:
    grads = len(Df_merge[Df_merge['education'].str.contains(uni) == True])
    gradsDict[uni] = grads
```

```python
avgDict = {}

for uni in unis:

    index = int(Df_rankings2[Df_rankings2['Name']== uni].index[0])
    numGrads = Df_rankings2.iloc[index]['Number of graduates']
    if numGrads == 0:
        continue
    Df_uni = Df_merge[Df_merge['education'].str.contains(uni) == True]
    
    rankSum = 0
    for index, row in Df_uni.iterrows():
        rank = row['ranking_final_b_5']
        rankSum += rank

    average = int(rankSum / numGrads)
    if average == 0:
        continue
    avgDict[uni] = average

Df_avgs = pd.DataFrame(list(avgDict.items()), columns=['Name', 'Average rank of graduates'])
Df_rankings2 = pd.merge(Df_rankings2, Df_avgs, on="Name")
Df_rankings2
```
